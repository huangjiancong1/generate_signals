```python
def feature_normalize(data):
    mu = np.mean(data,axis=0)
    std = np.std(data,axis=0)
    return (data - mu)/std

def un_normalize(normalized_data, input_data):
    mu = np.mean(input_data,axis=0)
    std = np.std(input_data,axis=0)
    return normalized_data*std+mu
```

```python
sides_3 = feature_normalize(sides_3)
```

```python
test_data = sides_3[0]
print(test_data)
```

```python
import torch.utils.data

test_loader = torch.from_numpy(test_data).cuda().float()
# test_loader=feature_normalize(test_loader)
print(test_loader)
```

```python
latent = ae.encoder(test_loader)
print(latent)
output = ae.decoder(latent).detach().cpu().numpy()

print (output)
```

```python
slides_3_generate = []
latent_space = []
sides_3 = feature_normalize(sides_3)

for test in sides_3:
    test_loader = torch.from_numpy(test).cuda().float()
    latent = ae.encoder(test_loader)
    output = ae.decoder(latent).detach().cpu().numpy()
    
    slides_3_generate.append(output)
    latent_space.append(latent.detach().cpu().numpy())
    

print (len(slides_3_generate))

np.save('slides_3_generate.npy',slides_3_generate)
np.save('latent_space.npy',latent_space)
```
